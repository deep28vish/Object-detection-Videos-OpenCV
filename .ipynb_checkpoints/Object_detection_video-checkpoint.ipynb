# Object Detection in Video- (OpenCV)

This is 2 of 3 notebooks check out the 1st notebook to learn about object detection and OpenCV for images- https://github.com/deep28vish/Object-Detection-Images-OpenCV

IMAGE detection in a video is same concept as image detection in images, as video is nothing but images flowing at 24 images per second.

Here again we will use classifiers by provided by OpenCV - https://github.com/opencv/opencv/tree/master/data/haarcascades

To avoid the clutter and make the output as smooth as possible we will use only 2 classifier in our vid.

Please Download example vid - https://www.vidsplay.com/peoplenyc.html

lets call in the most mportant library


```python
import cv2
```

Now we will be assigning Python variable to our video location. to make it simple keep the video in the same folder as our main .py .ipynb file as done in this repository.

We will also assign path so that our program knows where our classifiers are in the system.

```python
video = cv2.VideoCapture('peoplenyc.mp4')
a = 1
path = 'C:\\Users\\kuk\\OBJECT_DETECTION_VIDEO\\haar_cascade'
# path will contain the address of your xml. file locations
# a here is the number of frames that our program will process throughout
```

Now for the main part of the image detection program we will have a loop over our previous Image detection for images.


```python
# this snippet of code will have an ouput of dimension 500x500 in Gray scale 
# next snippet will have an output of 250x250 with color
while True:
    a = a+1 
    # check is a boolean value it remain True as ling as there are images coming in from our vid file
    # frame reads every frame(img) obtaine dfrom our vid
    check, frame = video.read()
    if check == False:
        break
    # for the peoplenyc.mp4 vid the image/ vid resolution is 1080x1920x3
    # which is very very high = 62,20,800 per frame and per second we have 
    # approx 25 frmaes and the vid is about 10seconds long that brings us to about 250images of high res.
    
    # as our classifiers depend on CPU capacity and will be completly exhausting CPU for image processing
    # we will see a very slow output vid as we are limited to our CPU cores.
    # NOTE: you dont really need 8 cores 16 threads to see the program works or not , even 2 core 
    # system will run the code but the output wont be as smooth as input
    # to reduce this pressure on the CPU we will convert our image to BLACK AND WHIT also called 
    # as Gray scale image and reduce the frame size to 500x500 from 1080x1920 thus further reducing the amount
    # of data flowing through CPU
    
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    gray_frame = cv2.resize(gray_frame, (500,500))
    
    # assigning xml file to our python variables along with the address there are physically present in
    face_cascade = cv2.CascadeClassifier(path +"\\haarcascade_frontalcatface.xml")
    fullbody_cascade = cv2.CascadeClassifier(path + '\\haarcascade_fullbody.xml')
    
    # using those variables now as classifiers to process the image and get the 
    # bounding box coordinates we need to display what was classified.
    faces = face_cascade.detectMultiScale(gray_frame, 1.05, 5)
    fullbody = fullbody_cascade.detectMultiScale(gray_frame, 1.05, 5)
    
    # after obtaining the values of bounding boxes we will now plot these boxes in or videos
    
    for x,y,w,h in faces:
        rec_img = cv2.rectangle(gray_frame, (x,y), (x+w, y+h), (0,255,0) ,0)
        face_text = cv2.putText(gray_frame, "FACE",(x, y+10),cv2.FONT_HERSHEY_COMPLEX_SMALL,.7,(0,255,0))
        
    for x,y,w,h in fullbody:
        rec_img = cv2.rectangle(gray_frame, (x,y), (x+w, y+h), (0,255,0) ,0)
        body_text = cv2.putText(gray_frame, "full body",(x, y+10),cv2.FONT_HERSHEY_COMPLEX_SMALL,.7,(0,255,0))
        
    # Displaying the output
    cv2.imshow('vid', gray_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
print('number of frames processed ',a)
video.release

cv2.destroyAllWindows()
```

The output will be somthing like this but a slow video


```python
# here we will repeat the same step but with color frames
video = cv2.VideoCapture('peoplenyc.mp4')
a = 1
while True:
    a = a+1 
    check, frame = video.read()
    if check == False:
        break

    frame = cv2.resize(frame, (250,250))
    
    face_cascade = cv2.CascadeClassifier(path +"\\haarcascade_frontalcatface.xml")
    fullbody_cascade = cv2.CascadeClassifier(path + '\\haarcascade_fullbody.xml')

    faces = face_cascade.detectMultiScale(frame, 1.05, 5)
    fullbody = fullbody_cascade.detectMultiScale(frame, 1.05, 5)
    
    for x,y,w,h in faces:
        rec_img = cv2.rectangle(frame, (x,y), (x+w, y+h), (0,255,0) ,0)
        face_text = cv2.putText(frame, "FACE",(x, y+10),cv2.FONT_HERSHEY_COMPLEX_SMALL,.7,(0,255,0))
        
    for x,y,w,h in fullbody:
        rec_img = cv2.rectangle(frame, (x,y), (x+w, y+h), (0,255,0) ,0)
        body_text = cv2.putText(frame, "full body",(x, y+10),cv2.FONT_HERSHEY_COMPLEX_SMALL,.7,(0,255,0))
        
  
    cv2.imshow('color vid', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

print('Number of color frmaes processed',a)
video.release

cv2.destroyAllWindows()
```
